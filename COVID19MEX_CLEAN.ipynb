```python
df1 = pd.read_excel(xls, sheets[1])
sector=df1.set_index('CLAVE').DESCRIPCIÓN.to_dict()
sector

df1 = pd.read_excel(xls, sheets[7])
entidad=df1.set_index('CLAVE_ENTIDAD').ENTIDAD_FEDERATIVA.to_dict()

df1 = pd.read_excel(xls, sheets[8])
municipio=df1.set_index('CLAVE_MUNICIPIO').MUNICIPIO.to_dict()

municipio
```

```python
origen={1:"USMER",
        2:"Fuera de USMER",
        99:"No Especificado"}

tipo_paciente={1:"Ambulatorio",
               2:"Hospitalizado",
               99:"No Especificado"}
               
sexo={1:"Mujer",
      2:"Hombre",
      99:"No especificado"}

si_no={1:"Si",
       2:"No",
       97:"No Aplica",
       98:"Se Ignora",
       99:"No Especificado"}

nacionalidad={1:"Mexicana",
              2:"Extranjera",
              99:"No Especificado"}

resultado={1:"Positivo SARS-CoV-2",
           2:"No positivo SARS-CoV-2",
           3:"Resultado pendiente"}

pais={"97":"No Aplica",
      "99":"Se Ignora"}
```

```python
covid19_df=covid19_df.replace({"ORIGEN":origen,
                               "PAIS_NACIONALIDAD":pais,
                               "TIPO_PACIENTE":tipo_paciente,
                               "INTUBADO":si_no,
                               "NEUMONIA":si_no,
                               "EMBARAZO":si_no,
                               "HABLA_LENGUA_INDIG":si_no,
                               "DIABETES":si_no,
                               "EPOC":si_no,
                               "ASMA":si_no,
                               "INMUSUPR":si_no,
                               "HIPERTENSION":si_no,
                               "OTRA_COM":si_no,
                               "CARDIOVASCULAR":si_no,
                               "OBESIDAD":si_no,
                               "RENAL_CRONICA":si_no,
                               "TABAQUISMO":si_no,
                               "MIGRANTE":si_no,
                               "PAIS_ORIGEN":pais,
                               "NACIONALIDAD":nacionalidad,
                               "UCI":si_no,
                               "RESULTADO":resultado,
                               "OTRO_CASO":si_no,
                               "SEXO":sexo,
                               "ENTIDAD_UM":entidad,
                               "SECTOR":sector,
                               "ENTIDAD_NAC":entidad,
                               "ENTIDAD_RES":entidad,
                               "MUNICIPIO_RES":municipio})

covid19_df.head()
```

```python
covid19_df['FECHA_ACTUALIZACION'] =pd.to_datetime(covid19_df.FECHA_ACTUALIZACION)
covid19_df['FECHA_INGRESO'] =pd.to_datetime(covid19_df.FECHA_INGRESO)
covid19_df['FECHA_SINTOMAS'] =pd.to_datetime(covid19_df.FECHA_SINTOMAS)
covid19_df['FECHA_DEF'] =pd.to_datetime(covid19_df.FECHA_DEF)
covid19_df.dtypes
```
